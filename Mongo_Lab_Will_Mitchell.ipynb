```python
host_name = "localhost"
port = "27017"

atlas_cluster_name = "cluster0.xuibg2h"
atlas_default_dbname = "sakila"
atlas_user_name = "williamhmitch2"
atlas_password = "Water2002"
```

```python
conn_str = {
    "local" : f"mongodb://{host_name}:{port}/",
    "atlas" : f"mongodb+srv://{atlas_user_name}:{atlas_password}@{atlas_cluster_name}.mongodb.net/{atlas_default_dbname}"
}

client = pymongo.MongoClient(conn_str["atlas"])

print(f"Local Connection String: {conn_str['local']}")
print(f"Atlas Connection String: {conn_str['atlas']}")
```

```python
db_name = "williamhmitch2_class"
db = client[db_name]
client.list_database_names()
```

```python
students = db["students"]
single_document = {
    "name": "Alvin",
    "race": "White",
    "occupation": "Lumberjack",
    "gender": "Male",
    "residence": "Apartment"

}
students.insert_one(single_document)
print("document inserted")

multiple_documents = [
    {"name": "Johnny", "race": "Asian", "occupation": "Stockbroker", "gender": "Male", "residence": "Home"},
    {"name": "Elizabeth", "race": "White", "occupation": "Teacher", "gender": "female", "residence": "Home"}
]
students.insert_many(multiple_documents)
print("multiple documents inserted")
```

```python
all_students = students.find()
print("all documents in students collection")
for student in all_students:
  print(student)

specific_student = students.find_one({"name": "Johnny"})
if specific_student:
  print("\ndocument with name'Johnny':")
  print(specific_student)
else:
  print("\nThere was no document with name 'Johnny'.")
```

```python
filter_query = {"name": "Elizabeth"}
update_data = {"$set": {"occupation": "Professor"}}

result = students.update_one(filter_query, update_data)

print("Number of documents matched correctly:", result.matched_count)
print("Number of documents modified correctly:", result.modified_count)
```

```python
filter_query = {"race": "white"}
update_data = {"$set": {"race": "hispanic"}}

result = students.update_many(filter_query, update_data)

print("\nNumber of documents matched correctly:", result.matched_count)
print("\nNumber of documents modified correctly:", result.modified_count)
```

```python
filter_query = {"name": "Alvin"}
result = students.delete_one(filter_query)

print("Number of documents deleted correctly:", result.deleted_count)
```

```python
filter_query = {"race": "hispanic"}
result = students.delete_many(filter_query)

print("Number of documents deleted correctly:", result.deleted_count)
```
